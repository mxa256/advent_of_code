```python
import pandas as pd

#Read the input
with open('input.txt', 'r') as file:
    content = list(file)

#Strip newline characters
content = [line.strip() for line in content]
```

```python
#Turn it into a dataframe 
content = pd.DataFrame(content, columns=['input'])
content['direction'] = content['input'].str[0]
content['value'] = content['input'].str[1:].astype(int)
content.drop('input', axis=1)
```

```python
#Initialize
pw_counter = 0
dial = 50

#test = ["L68", "L30", "R48", "L5", "R60", "L55", "L1", "L99", "R14", "L82"]
#test = pd.DataFrame(test, columns=['input'])
#test['direction'] = test['input'].str[0]
#test['value'] = test['input'].str[1:].astype(int)
#test.drop('input', axis=1)

#Let's see if this ugly for loop works
for direction, value in zip(content.direction, content.value):
    if direction == "R":
        dial = dial + value
        if dial > 99: 
            dial = (dial % 100)
            if dial == 0:
                pw_counter += 1
        elif dial < 0:
            dial = (dial % 100)
            if dial == 0:
                pw_counter += 1
        elif dial == 0:
            dial = dial
            pw_counter += 1
        else: 
            dial = dial 
    if direction == "L":
        dial = dial - value
        if dial > 99:
            dial = (dial % 100)
            if dial == 0:
                pw_counter += 1
        elif dial < 0:
             dial = (dial % 100)
             if dial == 0:
                 pw_counter +=1
        elif dial == 0:
            dial = dial
            pw_counter += 1
        else: 
            dial = dial 
```
